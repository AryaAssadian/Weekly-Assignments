```python
#General information given by the problem
T_inside=20  #degree C
T_outside=-10 # degre C
Total_area_of_the_wall=3*5             #m^2
area_of_the_considered_unit=1*0.25     #m^2

#We assume a depth of 1 meter to calculate Q unit. 

import pandas as pd
R_Name=["Inside air","Brick", "Plaster_1","Plaster_2","Plaster_3" ,"Plaster_4","Foam_1","Outside air"]
R_Conf=["series","Parallel","series","Parallel","Parallel","series","series","series"]
R_type=["Conv","Cond","Cond","Cond","Cond","Cond","Cond","Conv"]
Length=[None,0.16,0.02,0.16,0.16,0.02,0.03,None]           # m
K=[None,0.72,0.22,0.22,0.22,0.22,0.026,None]               # W/m^2.degree c
A=[0.25,0.22,0.25,0.015,0.015,0.25,0.25,0.25]              # m^2
h=[10,None,None,None,None,None,None,40]                    # W/m^2
R_Value=[0,0,0,0,0,0,0,0]
Resistance_list=[R_Conf,R_type,Length,K,A,h,R_Value]
Resistance_list_DF=pd.DataFrame(Resistance_list,("R_Conf","R_type","Length","K","A","h","R_Value"),R_Name)
Resistance_list_DF=Resistance_list_DF.transpose()
Resistance_list_DF
```

```python
#The aim of this section is to calculte the R_value of the materials.

iscond=Resistance_list_DF.loc[:,"R_type"]=="Cond"
Resistance_list_DF.loc[iscond,"R_Value"]=Resistance_list_DF["Length"]/(Resistance_list_DF["K"]*Resistance_list_DF["A"])

isconv=Resistance_list_DF.loc[:,"R_type"]=="Conv"
Resistance_list_DF.loc[isconv,"R_Value"]=1/(Resistance_list_DF["h"]*Resistance_list_DF["A"])
Resistance_list_DF
```

```python
#The aim of this section is to calculate the total resistance of the models. In order to do so, resistances in parallel and series are defined and calculated seperately.Then the total resistance is calculated as sum of the total Parallel and series resistances. 

R_series=Resistance_list_DF.loc[:,"R_Conf"]=="series"
Resistancelist_series=Resistance_list_DF.loc[R_series,:]
Total_Resistance_in_series=sum(Resistancelist_series.loc[:,"R_Value"])                   #degree C/W

R_Parallel=Resistance_list_DF.loc[:,"R_Conf"]=="Parallel"
Resistancelist_Parallel=Resistance_list_DF.loc[R_Parallel,:]
Total_Resistance_in_Parallel=1/(sum(1/Resistancelist_Parallel.loc[:,"R_Value"]))         #degree C/W

Total_Resistance_of_the_model=Total_Resistance_in_Parallel+Total_Resistance_in_series    #degree C/W



```

```python
#knowing all of the required parameres, Q dot may be calculated as follwos

Q_dot_unit=(T_inside-T_outside)/Total_Resistance_of_the_model          #W


Q_dot_Total=Q_dot_unit*(Total_area_of_the_wall/area_of_the_considered_unit)   #w
```
