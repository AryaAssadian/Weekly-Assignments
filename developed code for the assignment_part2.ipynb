```python
import pandas as pd
import numpy as np
import os
import matplotlib.pyplot as plt



#defining the path of the already prepared file

path_to_the_folder="/content/drive/MyDrive/EETBS 2021-2022/Processed Data"
file_name="DF_joined_arya_assadian.csv"
path_to_the_file=os.path.join(path_to_the_folder,file_name)

#reading the file
DF_joined=pd.read_csv(path_to_the_file, index_col=0, parse_dates=[0])
DF_joined

DF_joined_selected=DF_joined.loc["2014-07-1 00:00:00":"2014-07-1 23:00:00"]


#plotting

fig , ax =plt.subplots(3,1,sharex=True,dpi=150)

ax[0].plot(DF_joined_selected.index,DF_joined_selected.loc[:,["Irradiance"]])
ax[1].plot(DF_joined_selected.index,DF_joined_selected.loc[:,["Consumption"]])
ax[2].plot(DF_joined_selected.index,DF_joined_selected.loc[:,["temperature"]])

ax[0].legend(["Irradiance"])
ax[1].legend(["Consumption"])
ax[2].legend(["temperature"])


#saving the plot as a png file

figName= "Arya assadian_modified Irradiance, Consumption and temperature vs local hours.jpg"
path_to_folder_of_the_figures="/content/drive/MyDrive/EETBS 2021-2022/Processed Data/Figures"
path_to_figure=os.path.join(path_to_folder_of_the_figures,figName)
plt.savefig(path_to_figure)
```
